```python
import requests
from bs4 import BeautifulSoup
import os

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8'
}


os.makedirs('english_translation', exist_ok=True)
os.makedirs('audio', exist_ok=True)

# Dictionary to store countries for which we are getting both files
countries_with_both_files = {}

def fetch_and_save_translation(country_code, country_name):
    webpage_url = f"https://nationalanthems.info/{country_code.lower()}.htm"
    
    # Fetch the webpage content
    response = requests.get(webpage_url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Find the div with the specific title attribute
        translation_div = soup.find('div', {'title': ["English translation", "English lyrics"]})
        if translation_div:
            content_div_id = translation_div['id']
            content_div = soup.find('div', {'id': f'target-{content_div_id}'})
            if content_div:
                english_translation = content_div.get_text(strip=True)
                
                # Write the translation to a text file
                file_path = os.path.join('english_translation', f'{country_name}.txt')
                with open(file_path, 'w', encoding='utf-8') as file:
                    file.write(english_translation)
                print(f"English translation for {country_name} has been written to '{file_path}'")
                return True
            else:
                print(f"Content div not found for {country_name}.")
        else:
            print(f"Translation div not found for {country_name}.")
    else:
        print(f"Failed to retrieve {webpage_url}, status code: {response.status_code}")
    return False

def download_audio(country_code, country_name):
    mp3_link = f"https://nationalanthems.info/{country_code.lower()}.mp3"
    
    # Step 3: Download the audio file
    audio_response = requests.get(mp3_link, headers=headers)
    if audio_response.status_code == 200:
        file_name = os.path.join('audio', f'{country_name}.mp3')
        with open(file_name, 'wb') as file:
            file.write(audio_response.content)
        print(f'Audio file for {country_name} downloaded successfully: {file_name}')
        return True
    else:
        print(f'Failed to download audio file for {country_name}')
    return False

# Iterate through the top 110 country codes and names
for i, (country_code, country_name) in enumerate(counties.items()):
    if len(countries_with_both_files) >= 110:
        break
    translation_success = fetch_and_save_translation(country_code, country_name)
    audio_success = download_audio(country_code, country_name)
    if translation_success and audio_success:
        countries_with_both_files[country_code] = country_name

print("Countries with both files:", countries_with_both_files)
```

```python
import os
from PIL import Image
import seaborn as sns
import pandas as pd

import matplotlib.pyplot as plt

# Directory for PNG images
png_dir = 'flags'

# Get the list of files in the PNG directory
png_files = os.listdir(png_dir)

# Initialize a list to store aspect ratios
aspect_ratios = []

# Calculate aspect ratios for each flag image
for file in png_files:
    if file.endswith('.png'):
        img = Image.open(os.path.join(png_dir, file))
        width, height = img.size
        aspect_ratio = width / height
        aspect_ratios.append(aspect_ratio)

# Create a DataFrame for analysis
df = pd.DataFrame(aspect_ratios, columns=['Aspect Ratio'])

# Plot the distribution of aspect ratios
plt.figure(figsize=(10, 6))
sns.histplot(df['Aspect Ratio'], bins=20, kde=True)
plt.title('Distribution of Aspect Ratios of Flags')
plt.xlabel('Aspect Ratio')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate mean, median, and standard deviation of aspect ratios
mean_aspect_ratio = df['Aspect Ratio'].mean()
median_aspect_ratio = df['Aspect Ratio'].median()
std_aspect_ratio = df['Aspect Ratio'].std()

print(f"Mean Aspect Ratio: {mean_aspect_ratio}")
print(f"Median Aspect Ratio: {median_aspect_ratio}")
print(f"Standard Deviation of Aspect Ratios: {std_aspect_ratio}")

# Plot a boxplot to visualize the distribution of aspect ratios
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Aspect Ratio'])
plt.title('Boxplot of Aspect Ratios of Flags')
plt.xlabel('Aspect Ratio')
plt.show()
```

# Color Distribution: Extract the most dominant colors using clustering (e.g., K-means) or color histograms.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from PIL import Image
from sklearn.cluster import KMeans

# Function to extract dominant colors
def get_dominant_colors(image, k=5):
    image = image.convert('RGBA') 
    img_array = np.array(image)

    # Remove alpha channel if present
    if img_array.shape[-1] == 4:
        # Keep only non-transparent pixels
        img_array = img_array[img_array[:, :, 3] > 0][:, :3]
    else:
        img_array = img_array[:, :, :3]

    # Flatten the image array
    img_array = img_array.reshape(-1, 3)

    # Ensure there are enough pixels for KMeans clustering
    if len(img_array) < k:
        raise ValueError("Not enough pixels to cluster into {} colors.".format(k))

    # Apply K-means clustering
    kmeans = KMeans(n_clusters=k, n_init=10)
    kmeans.fit(img_array)

    # Get the cluster centers (dominant colors)
    colors = kmeans.cluster_centers_
    labels = kmeans.labels_

    # Count the number of pixels in each cluster
    counts = Counter(labels)

    # Sort colors by the number of pixels
    sorted_colors = [colors[i] for i in counts.keys()]
    sorted_counts = [counts[i] for i in counts.keys()]

    return sorted_colors, sorted_counts

# Directory for PNG images
png_dir = 'flags'

# Ensure the directory exists
if not os.path.exists(png_dir):
    raise FileNotFoundError(f"Directory '{png_dir}' not found. Check the path.")

# Get the list of PNG files
png_files = [f for f in os.listdir(png_dir) if f.endswith('.png')]

# Dictionary to store dominant colors for each flag
dominant_colors_dict = {}

# Extract dominant colors for each flag image
for file in png_files:
    img_path = os.path.join(png_dir, file)
    img = Image.open(img_path)
    try:
        dominant_colors, counts = get_dominant_colors(img, k=5)
        dominant_colors_dict[file] = dominant_colors
    except ValueError as e:
        print(f"Skipping {file}: {e}")

# Plot the dominant colors for a sample flag
if dominant_colors_dict:
    sample_flag = list(dominant_colors_dict.keys())[0]
    sample_colors = np.array(dominant_colors_dict[sample_flag]) / 255 

    plt.figure(figsize=(8, 2))
    plt.title(f'Dominant Colors in {sample_flag}')
    plt.imshow([sample_colors])
    plt.axis('off')
    plt.show()
else:
    print("No valid flag images found.")
```

# Code to Analyze Flag Resolutions

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from PIL import Image

# Directory containing PNG flag images
png_dir = "flags"

# Ensure the directory exists
if not os.path.exists(png_dir):
    raise FileNotFoundError(f"Directory '{png_dir}' not found. Check the path.")

# Get the list of PNG files
png_files = [f for f in os.listdir(png_dir) if f.endswith('.png')]

# List to store image resolutions
resolutions = []

# Extract width and height for each flag
for file in png_files:
    img_path = os.path.join(png_dir, file)
    img = Image.open(img_path)
    width, height = img.size
    resolutions.append((width, height))

# Count occurrences of each resolution
resolution_counts = Counter(resolutions)
most_common_resolutions = resolution_counts.most_common(10)

# Display the most common resolutions
print("\nTop 10 Most Common Flag Resolutions:")
for res, count in most_common_resolutions:
    print(f"Resolution: {res[0]}x{res[1]} - {count} flags")

# Convert resolutions to a NumPy array for plotting
res_array = np.array(resolutions)

# Plot histogram of flag widths and heights
plt.figure(figsize=(10, 5))
plt.hist(res_array[:, 0], bins=20, alpha=0.7, label="Width", color="blue")
plt.hist(res_array[:, 1], bins=20, alpha=0.7, label="Height", color="red")
plt.xlabel("Pixels")
plt.ylabel("Number of Flags")
plt.title("Distribution of Flag Resolutions")
plt.legend()
plt.show()
```

# Code for Shape Detection in Flags

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Directory containing flag images
png_dir = "flags"

# Ensure the directory exists
if not os.path.exists(png_dir):
    raise FileNotFoundError(f"Directory '{png_dir}' not found. Check the path.")

# Get list of flag images
png_files = [f for f in os.listdir(png_dir) if f.endswith('.png')]

# File to store shape detection results
results_file = "shapes_results.txt"
error_file = "errors.txt"

# Function to detect and label shapes in an image
def detect_shapes(image_path):
    # Load image
    image = cv2.imread(image_path)
    
    if image is None:
        return None, None  # Image loading failed

    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply Gaussian blur and edge detection
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    edges = cv2.Canny(blurred, 50, 150)

    # Find contours
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    shape_counts = {"Circle": 0, "Triangle": 0, "Rectangle": 0, "Polygon": 0, "Star": 0}

    for contour in contours:
        # Approximate the shape
        epsilon = 0.02 * cv2.arcLength(contour, True)
        approx = cv2.approxPolyDP(contour, epsilon, True)
        sides = len(approx)

        # Identify the shape based on the number of sides
        if sides == 3:
            shape_name = "Triangle"
        elif sides == 4:
            shape_name = "Rectangle"
        elif sides > 4:
            # Check if it's a circle
            area = cv2.contourArea(contour)
            perimeter = cv2.arcLength(contour, True)
            circularity = 4 * np.pi * (area / (perimeter ** 2)) if perimeter > 0 else 0

            if 0.75 < circularity < 1.25:
                shape_name = "Circle"
            else:
                shape_name = "Polygon"
        else:
            shape_name = "Star"

        shape_counts[shape_name] += 1

        # Draw the detected shape
        cv2.drawContours(image, [contour], -1, (0, 255, 0), 2)
        x, y, w, h = cv2.boundingRect(contour)
        cv2.putText(image, shape_name, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)

    return image, shape_counts

# Open files for storing results and errors
with open(results_file, "w") as results, open(error_file, "w") as errors:
    for file in png_files:
        img_path = os.path.join(png_dir, file)
        detected_image, shape_data = detect_shapes(img_path)

        if detected_image is None:
            errors.write(f"Failed to load: {file}\n")
            print(f"Skipping {file}: Image could not be loaded.")
            continue

        # Store shape data in the text file
        results.write(f"{file}: {shape_data}\n")
        print(f"\nShapes detected in {file}: {shape_data}")

        # Show the image with detected shapes
        plt.figure(figsize=(6, 4))
        plt.imshow(cv2.cvtColor(detected_image, cv2.COLOR_BGR2RGB))
        plt.title(f"Detected Shapes in {file}")
        plt.axis("off")
        plt.show()

print("\nShape analysis complete! Check 'shapes_results.txt' for results.")
print("Any failed images are logged in 'errors.txt'.")
```
